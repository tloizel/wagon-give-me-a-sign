```python
from google.cloud import bigquery
from google.oauth2 import service_account
import sys; sys.path.append('..')
from params import PROJECT_ID, TABLE_ID

def load_dataset(PROJECT_ID, TABLE_ID):

    print(PROJECT_ID, TABLE_ID)

    TABLE_ID = 'test_dataset.test10'

    credentials = service_account.Credentials.from_service_account_file("../bq_keys.json")
    #credentials = service_account.Credentials.from_service_account_file('~//Users//admin/code//tloizel//wagon-give-me-a-sign//bq_keys.json')

    client = bigquery.Client(project=PROJECT_ID, credentials=credentials)
    table = client.get_table(TABLE_ID)


    # Construct the SQL query to retrieve the table data
    query = f'SELECT * FROM `{table}`'

    # Submit the query and fetch the results
    df = client.query(query).to_dataframe()
    
    return df
```

```python
def search_best_model_params(X_train, y_train, X_test, y_test):
    thierry_model_LSTM_all = Sequential()

    thierry_model_LSTM_all.add(LSTM(128, return_sequences=True, activation='relu', input_shape=(60,1)))
    thierry_model_LSTM_all.add(LSTM(64, return_sequences=True, activation='relu'))
    thierry_model_LSTM_all.add(LSTM(32, return_sequences=True, activation='relu'))
    thierry_model_LSTM_all.add(LSTM(26, return_sequences=False, activation='relu'))
    #thierry_model_LSTM_all.add(Dropout(0.15))
    thierry_model_LSTM_all.add(Dense(26, activation='relu'))
    #thierry_model_LSTM_all.add(Dropout(0.3))
    thierry_model_LSTM_all.add(Dense(16, activation='relu'))
    thierry_model_LSTM_all.add(Dense(26, activation='softmax'))

    #model.compile(loss='mse', optimizer=Adam(learning_rate=0.5), metrics=['mae'])
    thierry_model_LSTM_all.compile(loss='sparse_categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])

    es = EarlyStopping(patience=7, restore_best_weights=True)

    #model.compile(loss='mse', optimizer=Adam(learning_rate=0.5), metrics=['mae'])
    thierry_model_LSTM_all.compile(loss='sparse_categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])

    es = EarlyStopping(patience=10, restore_best_weights=True)

    history = thierry_model_LSTM_all.fit(
        X_train, y_train, 
        #validation_data=(X_test, y_test),
        validation_split = 0.3,
        callbacks=[es],
        epochs=200, batch_size=32, verbose=0)

    params = thierry_model_LSTM_all.evaluate(X_test, y_test, verbose = 0)
    
    return params
    
```

# Sauvegarde du modèle

```python
import pickle

def upload_model(model, name_of_model):
    with open(f'{name_of_model}', 'wb') as file:
        pickle.dump(model, file)


def load_model(name_of_the_model=False):
    if name_of_the_model == False:
        with open('model.randomfo', 'rb') as file:
            loaded_model = pickle.load(file)
        return loaded_model
    else:
        with open(name_of_the_model, 'rb') as file:
            loaded_model = pickle.load(file)
        return loaded_model
```
