# Load du df depuis BQ

```python
from google.cloud import bigquery
from google.oauth2 import service_account
import sys; sys.path.append('..')
from params import PROJECT_ID, TABLE_ID

def load_dataset(PROJECT_ID, TABLE_ID):

    print(PROJECT_ID, TABLE_ID)

    TABLE_ID = 'test_dataset.test10'

    credentials = service_account.Credentials.from_service_account_file("../bq_keys.json")
    #credentials = service_account.Credentials.from_service_account_file('~//Users//admin/code//tloizel//wagon-give-me-a-sign//bq_keys.json')

    client = bigquery.Client(project=PROJECT_ID, credentials=credentials)
    table = client.get_table(TABLE_ID)


    # Construct the SQL query to retrieve the table data
    query = f'SELECT * FROM `{table}`'

    # Submit the query and fetch the results
    df = client.query(query).to_dataframe()
    
    return df
```

## Features

```python
X = df.drop(columns=['x_0', 'y_0', 'z_0', 'target'])
X.head(3)
```

## Formatage/encoding target y

```python
y = df['target']
y.head(3)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)
y_encoded
```

## Proportion/distribution des classes

```python
y.unique()
```

```python
# Dispersion des classes dans le DF d'origine
round(df[['target']].groupby(df['target']).count().rename(columns={'target': '%'}) / len(df['target']), 2)
```

## Modelisation

### Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=.30, random_state=1, stratify=y)
```

### Check proportion/distibution dans Train/test

```python
# Dispersion des classes dans y_train
y_train_temp = pd.DataFrame(y_train)
y_train_temp['%'] = 1
round(y_train_temp[['%']].groupby(y_train_temp[0]).count() / len(y_train_temp[['%']]), 2)
```

```python
# Dispersion des classes dans y_test
y_test_temp = pd.DataFrame(y_test)
y_test_temp['%'] = 1
round(y_test_temp[['%']].groupby(y_test_temp[0]).count() / len(y_test_temp[['%']]), 2)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from tensorflow.keras.callbacks import EarlyStopping
```

# RESULTAT DE CE MODELE : ACCURACY @ 100% et LOSS @ 0.004

# Pour le moment, calibré avec les lettres A, B et C uniquement

```python
thierry_model_LSTM = Sequential()

thierry_model_LSTM.add(LSTM(128, return_sequences=True, activation='relu', input_shape=(60,1)))
thierry_model_LSTM.add(LSTM(64, return_sequences=True, activation='relu'))
thierry_model_LSTM.add(LSTM(32, return_sequences=True, activation='relu'))
thierry_model_LSTM.add(LSTM(26, return_sequences=False, activation='relu'))
thierry_model_LSTM.add(Dense(26, activation='relu'))
thierry_model_LSTM.add(Dense(16, activation='relu'))
thierry_model_LSTM.add(Dense(3, activation='softmax'))

#model.compile(loss='mse', optimizer=Adam(learning_rate=0.5), metrics=['mae'])
thierry_model_LSTM.compile(loss='sparse_categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])

es = EarlyStopping(patience=7, restore_best_weights=True)

history = thierry_model_LSTM.fit(
    X_train, y_train, 
    #validation_data=(X_test, y_test),
    validation_split = 0.3,
    callbacks=[es],
    epochs=100, batch_size=16, verbose=1)

thierry_model_LSTM.evaluate(X_test, y_test, verbose = 1)
```

# Sauvegarde du modèle

```python
import pickle

def upload_model(model, name_of_model):
    with open(f'{name_of_model}', 'wb') as file:
        pickle.dump(model, file)


def load_model(name_of_the_model=False):
    if name_of_the_model == False:
        with open('model.randomfo', 'rb') as file:
            loaded_model = pickle.load(file)
        return loaded_model
    else:
        with open(name_of_the_model, 'rb') as file:
            loaded_model = pickle.load(file)
        return loaded_model
```

```python
upload_model(thierry_model_LSTM, 'thierry_model_LSTM')
```
